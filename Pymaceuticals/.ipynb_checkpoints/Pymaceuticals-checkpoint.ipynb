```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

df1 = pd.read_csv(mouse_drug_data_to_load)
df2 = pd.read_csv(clinical_trial_data_to_load)
combineddf = pd.merge(df1, df2, on="Mouse ID")

df = pd.DataFrame(combineddf)
df.head()
```

## Tumor Response to Treatment

```python
group_drug_time = df.groupby(['Drug', 'Timepoint'])
tumor_mean = group_drug_time['Tumor Volume (mm3)'].mean()
tumor_df = pd.DataFrame(tumor_mean)
tumor_df
```

```python
drug_time_error = group_drug_time['Tumor Volume (mm3)'].sem()
std_err_drugs = pd.DataFrame(drug_time_error)
std_err_drugs
```

```python
tumormunge = tumor_df.reset_index()
tumor_pivot = pd.pivot_table(tumormunge, index=['Timepoint'], columns = ['Drug'])
tumor_pivot
```

```python
# # Generate the Plot (with Error Bars)
tumor_capo = tumor_pivot['Tumor Volume (mm3)']['Capomulin']
tumor_ceft = tumor_pivot['Tumor Volume (mm3)']['Ceftamin']
tumor_inf = tumor_pivot['Tumor Volume (mm3)']['Infubinol']
tumor_keta = tumor_pivot['Tumor Volume (mm3)']['Ketapril']
x_axis = [0,5,10,15,20,25,30,35,40,45]

tum_plt_1 = plt.scatter(x_axis, tumor_capo, label="Capomulin")
tum_plt_2 = plt.scatter(x_axis, tumor_ceft, label ="Ceftamin")
tum_plt_3 = plt.scatter(x_axis, tumor_inf, label ="Infubinol")
tum_plt_4 = plt.scatter(x_axis, tumor_keta, label="Ketapril")
plt.legend(loc="best", fontsize="small", fancybox=True)
# # Save the Figure
plt.savefig("Output/tum_plt.png")
```

```python
# Show the Figure
plt.show()
```

![Tumor Response to Treatment](../Images/treatment.png)

## Metastatic Response to Treatment

```python
met_mean = group_drug_time['Metastatic Sites'].mean()
met_df = pd.DataFrame(met_mean)
met_df
```

```python
std_err_mets = group_drug_time['Metastatic Sites'].sem()
met_err_df = pd.DataFrame(std_err_mets)
met_err_df
```

```python
met_pivot = pd.pivot_table(met_df, index=['Timepoint'], columns = ['Drug'])
met_pivot
```

```python
# Generate the Plot (with Error Bars)
met_capo = met_pivot['Metastatic Sites']['Capomulin']
met_ceft = met_pivot['Metastatic Sites']['Ceftamin']
met_inf = met_pivot['Metastatic Sites']['Infubinol']
met_keta = met_pivot['Metastatic Sites']['Ketapril']

met_plt_1 = plt.scatter(x_axis, met_capo, label="Capomulin")
met_plt_2 = plt.scatter(x_axis, met_ceft, label ="Ceftamin")
met_plt_3 = plt.scatter(x_axis, met_inf, label ="Infubinol")
met_plt_4 = plt.scatter(x_axis, met_keta, label="Ketapril")

plt.legend(loc="best", fontsize="small", fancybox=True)

# Save the Figure
plt.savefig("Output/met_plt.png")

# Show the Figure
plt.show()
```
